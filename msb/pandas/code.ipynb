### 作业3  
使用题目二中的数据，进行数据筛选。

* 1、筛选索引大于 **'t'** 的所有数据
* 2、筛选学历是博士，身高大于170或者体重小于80的学生

```python
# 获取索引，进行逻辑运算
# data.index获取行索引
cond = data.index > 't'

data[cond]
```

```python
cond1 = data['学历'] == '博士'

cond2 = data['身高'] > 170

cond3 = data['体重'] < 80

cond = cond1 & (cond2 | cond3)

data[cond]
```

### 作业4  
使用题目二中数据，开始学生们开始减肥

* 本科生减肥，减掉的体重统一是10
* 博士生减肥，减掉体重范围是5~10

```python
cond = data['学历'] == '本科'

# 并没有改变原数据
# data[cond]['体重'] - 10

# location 定位，锁定，根据条件进行定位
# 一步到位，减法，然后赋值
data.loc[cond,'体重'] -= 10
```

```python
data[cond]
```

```python
cond = data['学历'] == '博士'
df2 = data[cond] # 复制了一份数据
df2.shape[0]
```

```python
# 在复制的数据基础上，进行了修改，源数据没有变
df2['体重'] - np.random.randint(5,11,size = df2.shape[0])
```

```python
# 原数据修改
cond = data['学历'] == '博士'

display(data[cond])

# data[cond].shape[0] 作用：获取多少名博士
data.loc[cond,'体重'] -= np.random.randint(5,11,size = data[cond].shape[0])

display(data[cond])
```

```python
cond = data['学历'] == '博士'
cond.sum() # False = 0,True = 1

display(data[cond])

data.loc[cond,'体重'] -= np.random.randint(5,11,size = cond.sum())

data[cond]
```
